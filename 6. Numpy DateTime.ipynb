```python
"""Creating datetime64 Objects"""
import numpy as np
date_only = np.datetime64('2025-08-31','D')
datetime_full = np.datetime64('2025-08-31T05:43:00')
print(f"Date only: {date_only}")
print(f"Full datetime: {datetime_full}")
```

```python
"""DateTime with specific units"""
year_unit = np.datetime64('2025-08-31', 'Y')
day_unit = np.datetime64('2025-08-31', 'D')
print(f"Year unit: {year_unit}")
print(f"Day unit: {day_unit}")
```

```python
"""Current date/time."""
current_time = np.datetime64('now')
today_date = np.datetime64('today', 'D')
print(f"Current time: {current_time}")
print(f"Today's date: {today_date}")
```

```python
"""Creating Arrays of datetime64"""
from datetime import datetime

dates_array = np.array(['2025-01-01', '2025-01-15', '2025-02-01'], dtype='datetime64')

python_datetime_list = [dt.item() for dt in dates_array]
print(f"Datetime array: {dates_array}")

print(python_datetime_list)
print(type(dates_array))
print(type(python_datetime_list))
```

```python
"""np.arange() for date ranges"""
month_dates = np.arange('2025-09', '2025-10', dtype='datetime64[D]')
print(f"Dates in August 2025: {month_dates}")
```

```python
"""Subtracting dates"""
date1 = np.datetime64('2026-10-31', 'Y')
date2 = np.datetime64('2025-08-01', 'Y')
difference = date1 - date2
print(f"Difference: {difference}")
```

```python
"""Addition/Subtraction with timedelta64"""
future_date = np.datetime64('2025-08-31') - np.timedelta64(14, 'D') #To know future dates or months or years/ past months, days or years
print(f"Date after 7 days: {future_date}")
```

```python
start_date = np.datetime64('2025-01-01') #To know future dates or months or years/ past months, days or years
date_after_30_days = start_date + np.timedelta64(30, 'D')
print(date_after_30_days)
```

```python
"""Extracting year, month, day"""
dt = np.datetime64('2025-08-31T10:30:00')
year = dt.astype('datetime64[Y]')
month = dt.astype('datetime64[M]')
day = dt.astype('datetime64[D]')
print(f"Year: {year}, Month: {month}, Day: {day}")
```

```python
nat_val = np.datetime64('nat') #Not a Time which indicates nan like in integer or not available
print(f"Not a Time: {nat_val}")
```

```python
# Today, yesterday, tomorrow
today = np.datetime64('today', 'D')
yesterday = today - np.timedelta64(1, 'D')
tomorrow = today + np.timedelta64(1, 'D')

# Dates in March 2023
march_2023_dates = np.arange('2023-03-01', '2023-04-01', dtype='datetime64[D]') #To generate dates/months/years use arange()
months = np.arange('2023-03-01', '2023-08-01', dtype='datetime64[M]')
#print(f"All dates of March 1st 2023 to April 1st 2023: {march_2023_dates}")
print(f"Months of 2023 required: {months}")
```

```python
date = np.datetime64('2025-08-31')
year = date.astype('datetime64[Y]')
month = date.astype('datetime64[M]')
day_of_week = (date - np.datetime64('2023-01-01')).astype('timedelta64[D]') #to calculate number of days present between the given dates. here we are converting to weeks
print(day_of_week)
```

```python
dates_in_march = np.arange('2017-03-01', '2017-04-01', dtype='datetime64[D]')
first_monday = dates_in_march[np.is_busday(dates_in_march, weekmask='Mon')][1] #Calculates first monday of the month for given month range
print(first_monday)
```
