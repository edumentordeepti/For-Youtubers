```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Load a sample dataset (or create your own DataFrame)
# Using Seaborn's built-in 'tips' dataset for demonstration
df = sns.load_dataset("tips")
df.head()
```

```python
# 2. Basic Scatter Plot
plt.figure(figsize=(8, 6)) # Set the figure size
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Total Bill vs. Tip (Basic Scatter Plot)")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True) # Add a grid for better readability
plt.show()
```

```python
# 3. Scatter Plot with Hue (coloring by a categorical variable)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", hue="time", data=df)
plt.title("Total Bill vs. Tip by Time of Day")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.show()
```

```python
# 4. Scatter Plot with Hue and Style (varying markers by another categorical variable)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", style="smoker", data=df)
plt.title("Total Bill vs. Tip by Day and Smoker Status")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.show()
```

```python
# 5. Scatter Plot with Size (varying point size by a numerical variable)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", size="size", data=df, sizes=(20, 400)) # Adjust size range
plt.title("Total Bill vs. Tip, Point Size by Party Size")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.show()
```

```python
# 6. Scatter Plot with Customizations (palette, markers, alpha)
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    style="day",
    size="size",
    data=df,
    palette="viridis", # Choose a color palette
    markers={"Thur": "o", "Fri": "s", "Sat": "P", "Sun": "X"}, # Custom markers for 'day'
    alpha=0.7, # Transparency of points
    edgecolor="w", # White edge color for points
    linewidth=0.5 # Line width for edges
)
plt.title("Customized Scatter Plot: Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.legend(title="Legend Details", bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend outside
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
