```python
#Subplotting design
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Load a sample dataset (or create one)
tips = sns.load_dataset("tips")

# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Various Seaborn Plots on Tips Dataset', fontsize=16)

# Plot 1: Scatter plot
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips, ax=axes[0, 0])
axes[0, 0].set_title("Total Bill vs Tip by Sex")

# Plot 2: Box plot
sns.boxplot(x="day", y="total_bill", hue="time", data=tips, ax=axes[0, 1])
axes[0, 1].set_title("Total Bill by Day and Time")

# Plot 3: Histogram
sns.histplot(x="tip", kde=True, data=tips, ax=axes[1, 0])
axes[1, 0].set_title("Distribution of Tip Amount")

# Plot 4: Count plot
sns.countplot(x="smoker", hue="day", data=tips, ax=axes[1, 1])
axes[1, 1].set_title("Smoker Count by Day")

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle
plt.show()
```

### Violin Plot: Similar to box plots but also shows the kernel density estimate of the distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
fmri = sns.load_dataset("fmri")

# Create a basic violin plot
sns.violinplot(x="timepoint", y="signal", data=fmri)

# Display the plot
plt.show()
```

```python
#Violin Plot
import numpy as np
import matplotlib.pyplot as plt

# Create sample data (uniform and normal distributions)
uniform_data = np.random.uniform(-100, 100, size=100)
normal_data = np.random.normal(size=100) * 30

# Combine data into a list for plotting
data_to_plot = [uniform_data, normal_data]

# Create a figure and axes
fig, ax = plt.subplots()

# Create the violin plot
ax.violinplot(data_to_plot, showmeans=False, showmedians=True)

# Add labels and title
ax.set_title("Uniform vs. Normal Data Distribution")
ax.set_ylabel("Observed Values")
ax.set_xticks([1, 2], ["Uniform", "Normal"]) # Set x-axis ticks for the two data sets

# Show the plot
plt.show()
```

```python
#Swarm Plot
#Similar to stripplot but arranges points to avoid overlap.

import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset (e.g., the 'tips' dataset)
tips = sns.load_dataset("tips")

# Create a basic swarm plot
# 'x' represents the categorical variable (e.g., 'day')
# 'y' represents the numerical variable (e.g., 'total_bill')
# 'data' specifies the DataFrame containing the variables
sns.swarmplot(x="day", y="total_bill", data=tips)

# Add title and labels for clarity
plt.title("Swarm Plot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")

# Display the plot
plt.show()

# Example with 'hue' for an additional categorical variable
plt.figure() # Create a new figure for the next plot
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips, palette="viridis")
plt.title("Swarm Plot of Total Bill by Day (Split by Sex)")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.show()
```

```python
#Point Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example dataset
tips = sns.load_dataset("tips")

# Create a basic point plot showing the average total_bill for each day
# and compare smokers vs. non-smokers
sns.pointplot(x="day", y="total_bill", hue="smoker", data=tips)

# Add titles and labels for clarity
plt.title("Average Total Bill by Day, for Smokers vs. Non-Smokers")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")

# Display the plot
plt.show()
```

```python
#stripplot
#shows the distribution of individual data points across a categorical variable
#and a continuous variable, using a scatter plot with added "jitter" (small random shifts) to prevent point overlap
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset("tips")

# Create the stripplot
# x: The categorical variable (day)
# y: The continuous variable (total_bill)
# data: The DataFrame containing the data
# jitter=True: Applies a small random displacement to prevent overlap of points
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)

# Add labels to the plot
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.title("Distribution of Total Bill by Day")

# Display the plot
plt.show()
```
