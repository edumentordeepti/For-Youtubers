```python
import numpy as np

# Define the two matrices as NumPy arrays
matrix_a = np.array([[1, 2, 3],
                     [4, 5, 6]])

matrix_b = np.array([[7, 8, 9],
                     [10, 11, 12]])

# Add the matrices using the + operator
result_matrix = matrix_a + matrix_b

# Print the resulting matrix
print(result_matrix)

result_matrix = np.add(matrix_a,matrix_b)

print(result_matrix)
```

```python
import numpy as np

# Define two 3D vectors
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

# Compute the cross product
cross_product_result = np.cross(vector_a, vector_b)

print(f"Vector A: {vector_a}")
print(f"Vector B: {vector_b}")
print(f"Cross Product: {cross_product_result}")
```

```python
from numpy import linalg
from numpy import linalg as la
determinant = la.det(matrix_a)
np.set_printoptions(precision=4)
print(f"The deteminant of matrix is:{determinant:.4f}")
```
