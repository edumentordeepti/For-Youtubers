1. autopct (Percentage Labels): Displays percentage values on each slice.
2. explode (Offsetting Slices): Emphasizes specific slices by moving them outward.
3. colors (Slice Colors): Assigns custom colors to each slice.
4. shadow (Shadow Effect): Adds a shadow beneath the pie chart for a 3D appearance.
5. startangle (Rotation): Rotates the starting point of the first slice.
6. wedgeprops (Wedge Properties): Customizes the appearance of the wedges (slices), such as linewidth and edgecolor.
7. radius (Chart Size): Adjusts the overall size of the pie chart.

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [35, 25, 20, 20]
explode = [0, 0.1, 0, 0.05] # Explode Bananas and Grapes slightly
colors = ['#FFDDC1', '#FFECB3', '#C5E1A5', '#81C784'] # Custom color palette
wedgeprops = {'linewidth': 0.8, 'edgecolor': 'grey'}

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140,
        wedgeprops = wedgeprops, radius=1.1)

plt.axis('equal') # Ensures circular aspect ratio
plt.title("Fruit Sales Distribution", fontsize=16)
plt.legend(labels, title="Fruits", loc="best", bbox_to_anchor=(1, 0.5)) # Add legend
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
data = np.random.randn(1000) * 5 + 10  # Normally distributed data

plt.figure(figsize=(10, 6)) # Custom figure size

# Create the histogram with various customizations
plt.hist(data,
         bins=30,
         color='lightgreen',
         edgecolor='darkgreen',
         alpha=0.8,
         density=False,
         label='Data Distribution')

# Add labels and title
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Customized Histogram of Sample Data', fontsize=14)

# Add a vertical line for the mean
mean_value = np.mean(data)
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')

# Add a legend
plt.legend()

# Add grid for better readability
plt.grid(axis='y', alpha=0.75)

plt.show()
```
