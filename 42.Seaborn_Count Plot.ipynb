```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load a built-in dataset for demonstration
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Create a simple countplot of the 'day' variable
plt.figure(figsize=(8, 6)) # Optional: set figure size
sns.countplot(x="day", data=tips)
plt.title("Count of Tips by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Tips")
plt.show()

# Create a countplot with 'hue' to further categorize by 'sex'
plt.figure(figsize=(8, 6))
sns.countplot(x="day", hue="sex", data=tips, palette="viridis")
plt.title("Count of Tips by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Tips")
plt.legend(title="Sex")
plt.show()

# Create a horizontal countplot
plt.figure(figsize=(8, 6))
sns.countplot(y="time", data=tips, palette="magma")
plt.title("Count of Tips by Time of Day")
plt.xlabel("Number of Tips")
plt.ylabel("Time of Day")
plt.show()

# Create a countplot with custom order for categories
plt.figure(figsize=(8, 6))
order_days = ["Thur", "Fri", "Sat", "Sun"]
sns.countplot(x="day", data=tips, order=order_days, palette="pastel")
plt.title("Count of Tips by Day (Custom Order)")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Tips")
plt.show()
```
