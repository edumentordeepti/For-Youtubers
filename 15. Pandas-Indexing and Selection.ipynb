#### Pandas DataFrames have an index that provides labels for rows. The set_index() and reset_index() methods are used to manipulate this index.

```python
import pandas as pd
```

```python
#The set_index() method allows you to change the DataFrame's index to one or more existing columns.
#data = pd.read_csv('multilingual_mobile_app_reviews_2025.csv')
data = pd.read_csv('multilingual_mobile_app_reviews_2025.csv', index_col="app_name")
print('Multilingual Mobile app reviews for 2025')
data.head()
```

```python
data.columns
```

```python
# Set 'review_id' as the new index
data_review_id = data.set_index('review_id')
print("\nDataFrame after setting 'review_id' as index:")
data_review_id.head(2)
```

```python
data_user_id = data.set_index('user_id')
print("\nDataFrame after setting 'user_id' as index:")
data_user_id.head(2)
```

```python
data.head(2)
```

#### 1) The column(s) chosen to be the new index are removed from the DataFrame's columns by default.
#### 2) You can create a multi-level index by passing a list of column names to set_index().
#### 3) Use inplace=True to modify the DataFrame directly without creating a new one.

```python
# Set new columns as MultiIndex
data_new_multiindex = data.set_index(['review_id', 'app_name'])
print("\nDataFrame after setting new MultiIndex:")
data_new_multiindex.head(2)
```

#### #reset_index():
1) The reset_index() method is used to reset the DataFrame's index to the default integer index (0, 1, 2, ...).
2) The current index can either be dropped or converted into a regular column.

```python
# Reset the index, making the old index a new column
data_reset1 = data_review_id.reset_index()
print("\nDataFrame after resetting index (old index as column):")
data_reset1.head(2)
```

```python

# Reset the index and drop the old index
data_reset2 = data_user_id.reset_index()
print("\nDataFrame after resetting index (old index dropped):")
data_reset2.head(2)
```

```python
# Reset the index and drop the old index
data_reset_dropped = data_review_id.reset_index(drop=True)
print("\nDataFrame after resetting index (old index dropped):")
data_reset_dropped.head(2)
```

#### 1) By default, reset_index() converts the old index into a new column named 'index'.
#### 2) Use drop=True to discard the old index instead of making it a column.
#### 3) Use inplace=True to modify the DataFrame directly.

```python
#Indexing Data using the [] Operator
#The [] operator is the basic and frequently used method for indexing in Pandas.
#It allows us to select columns and filter rows based on conditions.
#This method can be used to select individual columns or multiple columns.
"""Selecting a single column"""
data['app_name'].head()
```

```python
"""selecting multiple columns"""
data[['app_name', 'app_category']].head(5)
```

```python
data.head()
```

```python
"""Selecting Specific Rows and Columns"""
selection = data.loc[["MX Player", "Netflix"], ["user_id", "rating", "user_country", "user_gender"]]
print(selection.head())
```

```python
selection.shape
```

```python
"""Selecting All Rows and Specific Columns"""
all_rows_specific_columns = data.loc[:, ["user_id", "rating", "user_country", "user_gender"]]
print(all_rows_specific_columns.head())
```

```python
all_rows_specific_columns.shape
```

```python
"""Selecting single row by numerical position"""
row = data.iloc[3]
row.head()
```

```python
rows = data.iloc[[3, 5, 7],[0, 3, 4]]
rows.head()
```

```python
""" Access a single value for a row/column label pair"""
value = data.at['Netflix', 'review_text']
print(value)
```

```python
"""using query to index"""
result = data.query("review_id > 3 and user_gender == 'Female'")
result.head()
```

```python
"""Conditional Filtering"""
#Filtering by a Single Condition:
filtered_data = data[data['verified_purchase'] == False]
print("\nAll apps with verified purchase True")
filtered_data.head()
```

```python
filtered_data.shape
```

```python
"""Filtering by Multiple Conditions:"""
filtered_multiple = data[['app_category','rating']][(data['verified_purchase'] == True) & (data['user_country'] == 'India')]
filtered_multiple.head()
```
