```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('student.csv')
data.head(10)
```

#### Pandas rolling window functions allow you to perform calculations over a specified moving window of data within a Series or DataFrame.
#### This is particularly useful for time series analysis and smoothing data.

1) Define a window: You specify the size of the window (e.g., 3 days, 5 observations).
2) Slide the window: This window slides across your data, one observation at a time.
3) Apply a function: For each position of the window, a specified aggregation function (e.g., mean, sum, min, max, std) is applied to the data within that window.

```python
# Calculate a 3-period rolling mean
data['Rolling_Mean'] = data['mark'].rolling(window=3).mean()
print("DataFrame with Rolling Mean:")
data['Rolling_Mean'].head(20)
```

```python

# Calculate a 3-period rolling sum
data['Rolling_Sum'] = data['mark'].rolling(window=4).sum()
print("DataFrame with Rolling Sum:")
data['Rolling_Sum'].head(20)
```

```python
# Calculate a 3-period rolling max
data['Rolling_Max'] = data['mark'].rolling(window=3).max()
print("DataFrame with Rolling Max:")
data['Rolling_Max'].head(20)
```

```python
# Calculate expanding sum
data['expanding_sum'] = data['mark'].expanding().sum()
data['expanding_sum'].head(10)
```

```python
# Calculate expanding mean
data['expanding_mean'] = data['mark'].expanding().mean()
data['expanding_mean'].head(10)
```

```python
# Calculate expanding minimum
data['expanding_min'] = data['mark'].expanding().min()
data['expanding_min'].head(10)
```

```python
# Calculate expanding maximum
data['expanding_max'] = data['mark'].expanding().max()
data['expanding_max'].head(10)
```

```python
"""The moving average, also known as the rolling mean,
helps reduce noise and highlight significant patterns by averaging data points over a specific window."""
data['EMA_3'] = data['mark'].ewm(span=3, adjust=True).mean() #3-period moving average
data['EMA_3'].head()
```
