```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load an example dataset
# The 'tips' dataset is commonly used for demonstration purposes.
df = sns.load_dataset("tips")
df.head()
```

```python
# Set a style for the plot (optional, but improves aesthetics)
sns.set_style("whitegrid")

# Create a box plot
# x: categorical variable (e.g., 'day' of the week)
# y: quantitative variable (e.g., 'total_bill')
# data: the DataFrame containing the variables
plt.figure(figsize=(8, 6)) # Adjust figure size for better readability
sns.boxplot(x="day", y="total_bill", data=df)

# Add a title and labels for clarity
plt.title("Total Bill Distribution by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")

# Display the plot
plt.show()
```

```python
# Example with a hue variable for further grouping
plt.figure(figsize=(10, 7))
sns.boxplot(x="day", y="total_bill", hue="sex", data=df, palette="viridis")
plt.title("Total Bill Distribution by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.legend(title="Gender")
plt.show()
```

```python

# Example with custom whisker length and outlier styling
plt.figure(figsize=(8, 6))
sns.boxplot(x="time", y="tip", data=df, whis=1.5, fliersize=8, color="skyblue")
plt.title("Tip Distribution by Time of Day (Custom Whiskers & Outliers)")
plt.xlabel("Time of Day")
plt.ylabel("Tip Amount ($)")
plt.show()
```
