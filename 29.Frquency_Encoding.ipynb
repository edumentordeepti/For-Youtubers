#### Frequency encoding, also known as count encoding, replaces each category in a categorical feature with the number of times it appears in the dataset.
#### This transforms categorical data into a numerical format, which can be useful for machine learning models that require numerical inputs.

```python
#Frequency Encoding comes into picture when there is very heavy categorical values prsent in a dataset.

import pandas as pd

# Create a sample DataFrame
data = {'City': ['New York', 'London', 'Paris', 'New York', 'London', 'Tokyo', 'New York', 'Paris']}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Calculate the frequency of each city
city_counts = df['City'].value_counts()

# Map the frequencies back to the 'City' column
df['City_Encoded'] = df['City'].map(city_counts)

print("\nDataFrame after Frequency Encoding:")
print(df)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
```

```python
# fitting encoder

oe.fit_transform(df)
```
