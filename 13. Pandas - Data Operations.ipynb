```python
#Returns the first n rows of the DataFrame
"""Reading data from CSV"""
import pandas as pd

# Read a CSV file into a DataFrame
df_csv = pd.read_csv('Perfumes_dataset.csv')

# Display the first few rows of the DataFrame
print("DataFrame from CSV:")
print(df_csv.head(10))
#print(df_csv.head(n = 10))
```

```python
#Returns the last n rows of the DataFrame (default is 5).
print(df_csv.tail(10))
#print(df_csv.tail(n = 2))
```

```python
#info(): Provides a concise summary of the DataFrame, including data types, non-null values, and memory usage.
print(df_csv.info())
```

```python
#describe(): Generates descriptive statistics of numerical columns, including count, mean, standard deviation, min, max, and quartiles.
df_csv.describe()

```

```python
import pandas as pd

# Read a CSV file into a DataFrame
df_csv = pd.read_csv('student.csv')

# Display the first few rows of the DataFrame
print("DataFrame from CSV:")
print(df_csv.head(10))
#print(df_csv.head(n = 10))
```

```python
df_csv.describe(include='all')
```

```python
#shape: Returns a tuple representing the dimensions (rows, columns) of the DataFrame.
print(df_csv.shape)
```

```python
#columns: Returns an Index object containing the column labels of the DataFrame.
print(df_csv.columns)
```

```python
#index: Returns the Index object representing the row labels of the DataFrame.
print(df_csv.index)
```

```python
"""Display Options:
pd.set_option('display.max_rows', None): Displays all rows without truncation.
pd.set_option('display.max_columns', None): Displays all columns without truncation.
pd.set_option('display.width', 1000): Sets the display width to prevent line wrapping for wide DataFrames.
"""
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
print(df_csv)
```
