```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load a sample dataset
df = sns.load_dataset("penguins")

# Basic Jointplot with default scatter kind
print("Basic Jointplot (scatter):")
sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm")
plt.show()

# Jointplot with kind="kde" (Kernel Density Estimate)
print("\nJointplot with kind='kde':")
sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", kind="kde", fill=True, cmap="viridis")
plt.show()

# Jointplot with kind="reg" (Regression line)
print("\nJointplot with kind='reg':")
sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", kind="reg", line_kws={"color": "red"})
plt.show()

# Jointplot with kind="hex" (Hexagonal bins)
print("\nJointplot with kind='hex':")
sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", kind="hex", cmap="Blues")
plt.show()

# Customizing a scatter jointplot with markers, size, and marginal plot options
print("\nCustomized Scatter Jointplot:")
sns.jointplot(
    data=df,
    x="bill_length_mm",
    y="bill_depth_mm",
    marker="o",  # Change marker style
    s=70,        # Change marker size
    color="purple", # Change overall plot color
    marginal_kws=dict(bins=20, fill=False, color="green", kde=True), # Customize marginal plots
    joint_kws=dict(alpha=0.6) # Customize joint plot
)
plt.show()

# Adding more layers to a JointGrid object
print("\nJointGrid with additional layers:")
g = sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", kind="scatter", color="gray")
g.plot_joint(sns.kdeplot, color="red", zorder=0, levels=5) # Add KDE to joint plot
g.plot_marginals(sns.rugplot, color="blue", height=0.1) # Add rug plot to marginals
plt.show()
```
