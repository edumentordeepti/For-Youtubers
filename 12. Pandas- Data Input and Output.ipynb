```python
"""Reading data from CSV"""
import pandas as pd

# Read a CSV file into a DataFrame
df_csv = pd.read_csv('student.csv')

# Display the first few rows of the DataFrame
print("DataFrame from CSV:")
print(df_csv.tail())
```

```python
"""Reading data from Excel"""
import pandas as pd

# Read an Excel file into a DataFrame
df_excel = pd.read_excel('student.xlsx')

# Display the first few rows of the DataFrame
print("\nDataFrame from Excel:")
print(df_excel.head())
```

```python
"""reading fron dictionary"""
import pandas as pd

# Create a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Paris']
}

# Create a DataFrame from the dictionary
df_dict = pd.DataFrame(data)

# Display the DataFrame
print("\nDataFrame from Dictionary:")
print(df_dict)
```

```python
"""writing to CSV"""
import pandas as pd

# Sample DataFrame
data = {'Product': ['Laptop', 'Mouse', 'Keyboard'], 'Price': [1200, 25, 75]}
df_output = pd.DataFrame(data)

# Save DataFrame to CSV
df_output.to_csv("output.csv", index=False) # index=False prevents writing the DataFrame index as a column
print("\nDataFrame saved to output.csv")
print(df_output)
```

```python
"""writing to excel"""
import pandas as pd

# Sample DataFrame
data = {'Item': ['Shirt', 'Pants'], 'Size': ['M', 'L'], 'Stock': [50, 30]}
df_output_excel = pd.DataFrame(data)

# Save DataFrame to Excel
df_output_excel.to_excel("output.xlsx", index=False)
print("\nDataFrame saved to output.xlsx")
print(df_output_excel)
```

```python
"""writing to html"""
import pandas as pd

# Sample DataFrame
data = {'Country': ['USA', 'Canada'], 'Capital': ['Washington D.C.', 'Ottawa']}
df_output_html = pd.DataFrame(data)

# Save DataFrame to HTML
df_output_html.to_html("output.html", index=False)
print("\nDataFrame saved to output.html")
print(df_output_html)
```
