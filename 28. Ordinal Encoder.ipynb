```python
import pandas as pd

# Create a DataFrame with an ordinal categorical column
data = {'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Grade': ['C', 'A', 'B', 'A', 'C']}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Define the ordinal mapping
grade_mapper = {'C': 0, 'B': 1, 'A': 2}

# Apply the mapping to create a new encoded column
df['Grade_Encoded'] = df['Grade'].map(grade_mapper)

print("\nDataFrame after direct mapping ordinal encoding:")
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Create a DataFrame with an ordinal categorical column
data = {'Product': ['Small', 'Medium', 'Large', 'Small', 'Medium'],
        'Rating': ['Low', 'Medium', 'High', 'Low', 'Medium']}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Initialize OrdinalEncoder for 'Product' column
# Specify the order of categories explicitly
product_encoder = OrdinalEncoder(categories=[['Small', 'Medium', 'Large']])
df['Product_Encoded'] = product_encoder.fit_transform(df[['Product']])

# Initialize OrdinalEncoder for 'Rating' column
# Specify the order of categories explicitly
rating_encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])
df['Rating_Encoded'] = rating_encoder.fit_transform(df[['Rating']])

print("\nDataFrame after scikit-learn OrdinalEncoder:")
print(df)
```
