```python
"""The IQR is calculated as: IQR=Q3Q1 = Q3 - Q1
Data points that fall below Q11.5IQR or above Q3+1.5IQR are considered outliers."""
sort_data = np.sort(data[['Life_Expectancy','Spending_USD']])
sort_data
```

```python
Q1 = np.percentile(sort_data, 25, interpolation = 'midpoint')
Q2 = np.percentile(sort_data, 50, interpolation = 'midpoint')
Q3 = np.percentile(sort_data, 75, interpolation = 'midpoint')

print('Q1 25 percentile of the given data is, ', Q1)
print('Q2 50 percentile of the given data is, ', Q2)
print('Q3 75 percentile of the given data is, ', Q3)
```

```python
IQR = Q3 - Q1
print('Interquartile range is', IQR)
```

```python
low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
print('low_limit is', low_lim)
print('up_limit is', up_lim)
```

```python
sort_data
```

```python
data = pd.DataFrame(sort_data)
```

```python
data.head()
```

```python
outlier = []
for x in data[1]:
    if ((x> up_lim) or (x < low_lim)):
         outlier.append(x)
print(' outlier in the dataset')
outlier = pd.DataFrame(outlier)
outlier.count()
```

```python
#Using Stats IQR
import numpy as np
from scipy.stats import iqr

# Calculate Interquartile Range (IQR) using scipy
iqr_value = iqr(sort_data)

# Calculate Quartile Deviation
quartile_deviation = iqr_value / 2

print("Interquartile Range (IQR):", iqr_value)
print("Quartile Deviation:", quartile_deviation)
```

```python
lower_bound = Q1 - 1.5 * iqr_value
upper_bound = Q3 + 1.5 * iqr_value
print(lower_bound)
print(upper_bound)
```

```python
outlier = []
for x in data[1]:
    if ((x> upper_bound) or (x < lower_bound)):
         outlier.append(x)
print(' outlier in the dataset')
outlier = pd.DataFrame(outlier)
outlier
```

```python
outliers = sort_data[(sort_data < lower_bound) | (sort_data > upper_bound)]
print("Outliers:", outliers)
```
