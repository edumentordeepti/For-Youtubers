```python
#1. Line Plots
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = np.linspace(0, 10, 100) # Generate 100 points between 0 and 10
y = np.linspace(0, 20, 100)

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.show()
```

```python
#2. Scatter Plots
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50) # 50 random x-values
y = np.random.rand(50) # 50 random y-values

plt.scatter(x, y)
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Scatter Plot")
plt.show()
```

```python
#3. Bar Charts:
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Simple Bar Chart")
plt.show()
```

```python
#4. Histograms:
"""Histograms illustrate the distribution of a single numerical
variable by dividing data into bins and showing the frequency of values within each bin."""
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000) # 1000 random numbers from a standard normal distribution

plt.hist(data, bins=30) # Divide data into 30 bins
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Simple Histogram")
plt.show()
```

```python
#5. Pie Charts:
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.2f%%', startangle=90)
plt.axis('equal') # Ensures the pie chart is circular
plt.title("Simple Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title("Box Plot of Data Groups")
plt.ylabel("Value")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = [1, 5, 2, 8, 4]
plt.fill_between(x, y)
plt.title("Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
