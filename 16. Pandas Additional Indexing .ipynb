```python
import pandas as pd
data = pd.read_csv('multilingual_mobile_app_reviews_2025.csv')
#data = pd.read_csv('multilingual_mobile_app_reviews_2025.csv', index_col='app_name')
print('Multilingual Mobile app reviews for 2025')
data.head()
```

```python
data.shape
```

```python
"""iat- Access a single value for a row/column pair by integer position."""
# Access a single value using iat (row index 1, column index 2)
value = data.iat[1, 2]
print(f'\nValue at row 1, column 2: {value}')
```

```python
"""The dataframe.pop() method in Pandas removes a specified column from a DataFrame and returns it as a Pandas Series"""
# Use pop() to remove the 'review_id' column
# The removed column is returned as a Series
reviewid_column = data.pop('review_id')

print('\nDataFrame after pop():')
data.head()
```

```python
print("\nPopped 'review_id' column (as a Series):")
print(reviewid_column.head())
```

```python
"""Pandas dataframe.xs() function return a cross-section from the Series/DataFrame for the given key value."""
# return cross-section corresponding to
xsdata = data.xs(key = 'Netflix')
xsdata.head()
```

```python
xsdata.shape
```

```python
"""Pandas dataframe.get() function is used to get item from object for given key.
The key could be one or more than one dataframe column.
It returns default value if not found."""
# applying get() function
data.get('review_text').head()
```

```python
"""DataFrame.where() function replace values in a DataFrame based on a condition.
 It allows you to keep the original value where a condition is True and replace it with something else
  e.g., NaN or a custom value where the condition is False."""

res_True = data[][data['verified_purchase'].where(data['device_type'] == 'Windows Phone') == True]
#res_False = data['verified_purchase'][data['verified_purchase'].where(data['device_type'] == 'Windows Phone') == False]
print(res_True)
#print(res_False)
```

```python
"""Pandas dataframe.mask() function return an object of same shape as self
and whose corresponding entries are from self where cond is False and otherwise are from other object."""
# replace values greater than 10 with -25
data['rating'].mask(data['rating'] == 5 , 4.5).head(20)
```
