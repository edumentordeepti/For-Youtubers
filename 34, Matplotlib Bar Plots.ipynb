```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

plt.figure(figsize=(8, 6)) # Set figure size

# Basic bar plot with customization
plt.bar(
    categories,
    values,
    color=['yellow'], # Individual bar colors
    edgecolor='black', # Bar edge color
    linewidth=1.5, # Edge line width
    width=0.7, # Bar width (default is 0.8)
    alpha=0.8 # Transparency
)

plt.title('Customized Bar Plot', fontsize=16)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)

# Customize tick labels
plt.xticks(rotation=45, ha='right', fontsize=10) # Rotate x-axis labels, align right
plt.yticks(fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7) # Add horizontal grid lines

plt.show()
```

```python
#Horizontal Bar Plot
import matplotlib.pyplot as plt
import numpy as np

categories = ['Category 1', 'Category 2', 'Category 3']
values = [20, 35, 18]

plt.figure(figsize=(8, 5))
plt.barh(categories, values, color='teal', edgecolor='black')
plt.title('Horizontal Bar Plot')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```
