1. s (size): Controls the size of the markers. It can be a single value for uniform size or an array for varying sizes per point.
2. c (color): Defines the color of the markers. This can be a single color string (e.g., 'red', 'blue'), an array of color strings, or a numerical array to be mapped to a colormap.
3. marker: Specifies the shape of the markers (e.g., 'o' for circles, '^' for triangles, 's' for squares, 'D' for diamonds).
alpha: Sets the transparency level of the markers, ranging from 0 (fully transparent) to 1 (fully opaque).
4. edgecolors: Determines the color of the marker edges.
linewidths: Controls the width of the marker edges.
5. cmap (colormap): Used in conjunction with c when c is a numerical array, defining the colormap to use (e.g., 'viridis', 'plasma').
norm: Normalizes the c data when a colormap is used.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
sizes = np.random.rand(50) * 500 + 50 # Varying sizes
colors = np.random.rand(50) # For colormap

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7,
            marker='D', edgecolors='blue', linewidths=0.8)
plt.colorbar(label='Color Value')
plt.title('Customized Scatter Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data_x = [1, 2, 3, 4, 5, 6, 7, 8]
data_y = [5, 7, 4, 8, 6, 9, 3, 7]
categories = ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B']
category_colors = {'A': 'red', 'B': 'blue', 'C': 'green'}
colors = [category_colors[cat] for cat in categories]

plt.figure(figsize=(7, 5))
plt.scatter(data_x, data_y, c=colors, s=100, alpha=0.8)
plt.title('Scatter Plot with Categorical Colors')
plt.xlabel('Data X')
plt.ylabel('Data Y')
plt.show()
```
