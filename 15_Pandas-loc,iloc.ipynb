#### loc and iloc are fundamental Pandas DataFrame indexers used for selecting data. loc is label-based, while iloc is integer-position based.
#### The loc() function is label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc().
#### The iloc() function is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column. This method does not include the last element of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc().

```python
#loc()-(Label-based indexing)
import pandas as pd

data = pd.read_csv('IMDB_processed_data.csv')

# Display the first few rows of the DataFrame
print("IMDB Table")
data
```

```python
index_col = ['one','two','three','four','five','six','seven','eight','nine','ten','eleven','twelve','thirteen','fourteen','fifteen',
         'sixteen','seventeen','eighteen','ninteen','twenty']
data_renamed_index = data.set_axis(index_col, axis=0)
data_renamed_index
```

```python
data
```

```python
#Selecting a single row by label
print(data_renamed_index.loc[['two']])
```

```python
#Selecting a single row by label
print(data.loc[[1]])
```

```python
#Selecting a single row by label
print(data_renamed_index.loc[['two','five','six','eight']])
```

```python
#Selecting multiple rows by labels
print(data.loc[[2, 4, 6, 8 ,9 ]])
```

```python
data[['Rank','Title']]
```

```python
data
```

```python
print(data.loc[(data.Title == 'The Shawshank Redemption') & (data.Ratings == 9.3)])
```

```python
#Selecting rows and specific columns by labels
print(data_renamed_index.loc[['two', 'four'], ['Title', 'Release']])
print(data.loc[[2, 4], ['Title', 'Release']])
```

```python
#Selecting rows based on a condition.
print(data.loc[data['Ratings'] > 9.0])
```

```python
#Modifying data using loc
data.loc[1, 'Title'] = 'The Scorpion King'
print(data.head(2))
```

```python
#iloc: Selecting a single row by integer position:
print(data.iloc[[2]])
```

```python
#Selecting multiple rows by integer positions.
print(data.iloc[[2, 4]])
```

```python
#Slicing rows and columns by integer positions:
print(data.iloc[0:2, 0:2]) # Selects first two rows and first two columns
```

```python
#Selecting all rows and specific columns by integer positions:
print(data.iloc[:, [0, 2]]) # Selects all rows, and the first and third columns
```

```python
#Modifying data using iloc.
data.iloc[0, 1] = 20
print(data.head(2))
```
