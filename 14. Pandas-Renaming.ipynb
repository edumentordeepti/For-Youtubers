```python
import pandas as pd
```

```python
data = pd.read_csv('Perfumes_dataset.csv')

# Display the first few rows of the DataFrame
print("Original Table")
print(data.head())
```

```python
data.columns
```

```python
# Rename specific columns using a dictionary
data_renamed_cols = data.rename(columns={'brand': 'perfume brand', 'perfume': 'perfume_name'}, inplace=True)
#data_renamed_cols = data.rename(columns={'brand': 'perfume brand', 'perfume': 'perfume_name'})
print("\nAfter renaming specific columns:")
print(data_renamed_cols.head())
```

```python
data.head()
```

```python
# Rename all columns by assigning a new list
print(data.columns)
data.columns = ['perfume brand', 'perfume_name', 'perfume_type','category','audience', 'longevity']
print("\nDataFrame after renaming all columns:")
print(data.head())
```

```python
# Rename specific index labels using a dictionary
#data_renamed_idx = data.rename(index={'1': 'One', '2': 'Two'},inplace=True)
data_renamed_idx = data.rename(index={1: 'One', 2: 'Two'})
print("\nDataFrame after renaming specific index labels:")
print(data_renamed_idx.head())
```

```python
"""Setting Row Labels (Index)"""
#set_axis() function assigns the required index to either a column or index axis.
# Set new row labels (index)
new_indexes= ['a', 'b', 'c']
#data_with_new_index = data.set_axis(new_indexes, axis='index',inplace=True)
data_with_new_index = data.set_axis(new_indexes, axis='index')
print("\nDataFrame with new row labels:")
print(data_with_new_index)
```

```python
# Set new column labels
new_columns = ['perfume brand', 'perfumr_name', 'perfumr_type','category','audience', 'longevity']
#data_with_new_columns = data.set_axis(new_columns, axis='columns',inplace=True)
#data_with_new_columns = data.set_axis(new_columns, axis='columns')
data_with_new_columns = data.set_axis(new_columns, axis='columns') #axis value can be 'rows'/'columns' or 0/1
print("\nDataFrame with new column labels:")
print(data_with_new_columns.head())
```

```python
"""add_prefix(), add_suffix() function"""
data_prefixed = data.add_prefix('col_')
print("\nAfter adding prefix 'col_':")
print(data_prefixed.head())
```

```python
data_suffixed = data.add_suffix('_col')
print("\nAfter adding suffix '_col':")
print(data_suffixed.head())
```
