```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load a sample dataset
iris = sns.load_dataset("iris")

# Create a basic pairplot
print("Basic Pairplot:")
sns.pairplot(iris)
plt.suptitle("Basic Pairplot of Iris Dataset", y=1.02) # Add a title
plt.show()

# Pairplot with hue for categorical variable
print("\nPairplot with Hue:")
sns.pairplot(iris, hue="species")
plt.suptitle("Pairplot of Iris Dataset by Species", y=1.02)
plt.show()

# Pairplot with different diagonal kind (KDE) and customized plot aesthetics
print("\nPairplot with KDE on Diagonal and Custom Aesthetics:")
sns.pairplot(iris,
             hue="species",
             diag_kind="kde",
             plot_kws={'alpha': 0.7, 's': 50, 'edgecolor': 'k'},
             diag_kws={'fill': True, 'alpha': 0.5})
plt.suptitle("Pairplot with KDE and Custom Aesthetics", y=1.02)
plt.show()

# Pairplot with selected variables
print("\nPairplot with Selected Variables:")
sns.pairplot(iris,
             vars=["sepal_length", "sepal_width"],
             hue="species",
             kind="reg") # Use regression plots for off-diagonal
plt.suptitle("Pairplot with Selected Variables and Regression Lines", y=1.02)
plt.show()

# Pairplot with specific x and y variables
print("\nPairplot with Specific X and Y Variables:")
sns.pairplot(iris,
             x_vars=["sepal_length", "petal_length"],
             y_vars=["sepal_width", "petal_width"],
             hue="species")
plt.suptitle("Pairplot with Specific X and Y Variables", y=1.02)
plt.show()
```
