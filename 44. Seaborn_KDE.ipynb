```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Generate some example data
# Create a DataFrame for better organization and potential use with 'hue'
data = {
    'Value': np.random.normal(loc=0, scale=1, size=500),
    'Category': np.random.choice(['A', 'B'], size=500)
}
df = pd.DataFrame(data)

# Bivariate KDE plot for two variables
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='Value', y='Value', fill=True, cmap='viridis')
plt.title('Bivariate KDE Plot of Value vs Value')
plt.xlabel('Value')
plt.ylabel('Value')
plt.show()

# KDE plot with 'hue' for categorical separation
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='Value', hue='Category', fill=True, palette='coolwarm')
plt.title('KDE Plot of Value by Category')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend(title='Category')
plt.show()

# 3. Further customization (optional)
# Adjusting bandwidth, cut, and clip
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='Value', bw_adjust=2, cut=0, clip=(-2, 2), fill=True, color='lightcoral')
plt.title('KDE Plot with Custom Bandwidth and Clip')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
