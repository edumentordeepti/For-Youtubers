```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
np.random.seed(10)
data = [np.random.normal(loc, scale, 200) for loc, scale in zip([100, 90, 80, 70], [10, 20, 30, 40])] #zip-->[(100,10),(90,20),(80,30),(70,40)]
labels = ['Group A', 'Group B', 'Group C', 'Group D']

#figure and axes
fig = plt.subplots(figsize=(10, 7))

# Create the boxplot with customization options
bp = plt.boxplot(data, patch_artist=True,  # Fill boxes with color
                notch=True,             # Add notches to boxes
                vert=True,              # Vertical orientation
                labels=labels,          # Set x-axis labels
                showmeans=True          # Show mean as a point
               )

# Define custom colors for the boxes
colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightgoldenrodyellow']

# Customize boxes
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_edgecolor('black')
    patch.set_linewidth(1.5)

# Customize whiskers #Whiskers are tails of box plot
for whisker in bp['whiskers']:
    whisker.set(color='darkblue', linewidth=1.5, linestyle=':')

# Customize caps
for cap in bp['caps']: #Caps are top of whiskers
    cap.set(color='darkblue', linewidth=2)

# Customize medians #median of box plot
for median in bp['medians']:
    median.set(color='red', linewidth=3)

# Customize fliers (outliers) #outside whiskers all circular presentations are whiskers
for flier in bp['fliers']:
    flier.set(marker='o', color='purple', alpha=0.6, markersize=8)

# Customize mean points #mean line of box plot
for mean in bp['means']:
    mean.set(marker='^', markerfacecolor='green', markeredgecolor='green',
             markersize=10, linestyle='None')

# Add title and labels
plt.title('Customized Box Plot Example', fontsize=16)
plt.xlabel('Groups', fontsize=12)
plt.ylabel('Values', fontsize=12)

# Customize tick parameters
plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=10)

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
