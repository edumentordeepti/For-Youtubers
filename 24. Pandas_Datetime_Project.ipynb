```python
data['Date Posted'] = pd.to_datetime(data['Date Posted'])
```

```python
data['Date Posted'].dtype
```

```python
#Extract Time Components (if needed for grouping):
# Extract month and quarter for grouping
data['day'] = data['Date Posted'].dt.day
data['month'] = data['Date Posted'].dt.month
data['quarter'] = data['Date Posted'].dt.quarter
```

```python
#daily likes or per day likes for posts
daily_output = data.groupby(data['Day'])['Likes'].sum()
daily_output
```

```python
#Daily Output
date_indexed = data.set_index('Date Posted')
daily_resampled = date_indexed['Likes'].resample('D').sum()
daily_resampled
```

```python
#Monthly output
date_indexed = data.set_index('Date Posted')
monthly_resampled = date_indexed['Numer of Comments'].resample('M').sum()
monthly_resampled
```

```python
#Quaterly Output
quarterly_output = data.groupby('quarter')['Likes'].sum()
quarterly_output.head()
```

```python
# Alternative using resample
quarterly_resampled = date_indexed['Likes'].resample('Q').sum()
quarterly_resampled
```

```python
# Alternative using resample
yearly_resampled = date_indexed['Likes'].resample('Y').sum()
yearly_resampled
```
