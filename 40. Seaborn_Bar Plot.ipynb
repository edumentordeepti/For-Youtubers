```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Load a sample dataset (or create one)
# Using Seaborn's built-in 'tips' dataset for demonstration
df = sns.load_dataset("tips")
df.head()
```

```python
# 2. Set the aesthetic style of the plots
sns.set_theme(style="whitegrid")

# 3. Create the bar plot
# x: categorical variable, y: numerical variable
# hue: another categorical variable for grouping within bars
# data: the DataFrame containing the data
# estimator: function to aggregate numerical values (default is mean)
# ci: confidence interval size (e.g., "sd" for standard deviation, None to remove)
# palette: color palette for the bars
plt.figure(figsize=(10, 6)) # Set the figure size for better readability


sns.barplot(x="day", y="total_bill", hue="sex", data=df, estimator=sum, ci=None, palette="viridis")

# 4. Customize the plot (optional)
plt.title("Total Bill Amount by Day and Gender", fontsize=16)
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Total Bill Amount (Sum)", fontsize=12)
plt.legend(title="Gender") # Customize legend title

# 5. Display the plot
plt.show()
```
