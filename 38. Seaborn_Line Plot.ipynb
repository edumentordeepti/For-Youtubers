```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Load data (using a built-in Seaborn dataset for simplicity)
# You can also load your own data using pd.read_csv(), pd.read_excel(), etc.
flights_data = sns.load_dataset("flights")

# Display the first few rows of the dataset to understand its structure
print("Flights Data Head:")
print(flights_data.head())

# 2. Create a basic line plot
# Plotting passengers over the years
plt.figure(figsize=(10, 6)) # Set the figure size for better readability
sns.lineplot(data=flights_data, x="year", y="passengers")
plt.title("Total Airline Passengers Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.grid(True) # Add a grid for easier reading of values
plt.show()

# 3. Create a more customized line plot with hue and style
# Plotting passengers over months, separated by year
plt.figure(figsize=(12, 7))
sns.lineplot(data=flights_data, x="month", y="passengers", hue="year", style="year", marker="o")
plt.title("Monthly Airline Passengers by Year")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")
plt.xticks(rotation=45) # Rotate x-axis labels if they overlap
plt.legend(title="Year", bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside the plot
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

# 4. Example with a custom dataset
# Create a sample DataFrame
data = {
    'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
                            '2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']),
    'Value': [10, 12, 15, 13, 16, 8, 11, 14, 12, 15],
    'Category': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B']
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Date", y="Value", hue="Category", palette="viridis")
plt.title("Value Over Time by Category")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```
