```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Generate Sample Data
# Create a random 2D array for the heatmap
data = np.random.rand(10, 8)

# Create a DataFrame for better labeling and potential use of categorical data
df = pd.DataFrame(data,
                  columns=[f'Feature_{i+1}' for i in range(8)],
                  index=[f'Sample_{i+1}' for i in range(10)])

# 2. Create the Heatmap
plt.figure(figsize=(10, 7)) # Set the figure size for better readability
ax = sns.heatmap(df,
                 annot=True,      # Show the values in each cell
                 fmt=".2f",       # Format the annotations to two decimal places
                 cmap="viridis",  # Choose a colormap (e.g., "viridis", "coolwarm", "Blues")
                 linewidths=.5,   # Add lines between cells
                 linecolor='black', # Color of the lines between cells
                 cbar_kws={'label': 'Value Intensity'}, # Add a label to the color bar
                 center=0.5,     # Center the colormap around a specific value (useful for diverging data)
                 vmin = -1
                 )

# 3. Customize the Plot
ax.set_title("Seaborn Heatmap Example", fontsize=16) # Add a title to the heatmap
ax.set_xlabel("Features", fontsize=12) # Label the x-axis
ax.set_ylabel("Samples", fontsize=12) # Label the y-axis

plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better visibility
plt.yticks(rotation=0)             # Ensure y-axis labels are horizontal

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
