```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(0, 10, 0.1)
y1 = 2 * x + 5 + np.random.randn(len(x)) * 5
y2 = 0.5 * x**2 - 3 * x + 10 + np.random.randn(len(x)) * 5

# Create the figure and axes
fig = plt.subplots(figsize=(10, 6))

# Plot the area between y1 and y2
plt.fill_between(x, y1, y2,
                where=(y1 > y2), # Fill only where y1 is greater than y2
                color='skyblue',
                alpha=0.6,
                label='Y1 > Y2')

plt.fill_between(x, y1, y2,
                where=(y1 <= y2), # Fill only where y2 is greater than or equal to y1
                color='lightcoral',
                alpha=0.6,
                label='Y1 <= Y2')

# Plot the individual lines for clarity
plt.plot(x, y1, color='blue', linestyle='-', linewidth=1.5, label='Series Y1')
plt.plot(x, y2, color='red', linestyle='--', linewidth=1.5, label='Series Y2')

# Customize the plot
plt.title('Customized Area Plot Example', fontsize=16, fontweight='bold')
plt.xlabel('X-axis Label', fontsize=12)
plt.ylabel('Y-axis Label', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(loc='upper left', fontsize=10)
#plt.facecolor('#f0f0f0') # Set background color of the plot area
#fig.patch.facecolor('white') # Set background color of the entire figure

# Add text annotation
plt.text(7, 40, 'Peak Region', fontsize=10, color='darkgreen',
        bbox=dict(facecolor='white', alpha=0.7, edgecolor='green', boxstyle='round,pad=0.5'))

# Set x and y limits
plt.xlim(0, 10)
plt.ylim(0, 70)

# Show the plot
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
