```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('insurance.csv')
data.head()
```

```python
data.info()
```

#### Pandas DataFrames offer built-in plotting functionalities, leveraging Matplotlib, to facilitate basic data visualization directly from the DataFrame object.
#Valid plot kinds: ('line', 'bar', 'barh', 'kde', 'density', 'area', 'hist', 'box', 'pie', 'scatter', 'hexbin')

```python
# Create a sample DataFrame
#Line Plot
data.plot.line()
```

```python
#Bar Plots.
#Suitable for comparing categorical data or showing the distribution of a categorical variable.
data[['age','bmi']][:20].plot(kind='bar')
```

```python
#histograms.
#Used to visualize the distribution of a continuous variable by dividing values into bins.
data[['age','bmi']][:20].plot(kind='hist', bins=10)
```

```python
#Scatter Plots.
#Useful for examining the relationship between two continuous variables.
data[:20].plot.scatter(x='age',y='bmi', title='Age vs. BMI', grid=True)
```

```python
#Box Plots.
#Provide a summary of the distribution of a numerical variable, including median, quartiles, and outliers.
data[['age','bmi']].plot(kind='box')
#df.plot.box()
```

```python
#Area Plots.
#Show the cumulative distribution of a variable over a continuous or categorical axis, filling the space below the line.
data[['age','bmi']][:20].plot(kind='area', stacked=True) # stacked=True for cumulative areas
```
